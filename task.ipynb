# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')

df.head()
```

# Solutions

### 1. What are the column names in the dataset?

```python
cols = df.columns
cols
```

### 2. What is the average rating of the movies?

```python
average_rating = df['IMDB_Rating'].mean()

average_rating
```

### 3. How many unique genres are there?

```python
unique_genres = df['Genre'].nunique()

unique_genres
```

### 4. Save your results as a csv file.

```python
results = pd.DataFrame({'Question': ['Column Names', 'Average Rating', 'Number of Unique Genres'],
                       'Answer': [list(cols), average_rating, unique_genres]})
results.to_csv('results.csv', index=False)
rs = pd.read_csv('results.csv')
rs.head()
```

### 5. Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.

```python
# plot histogram for genre distribution
plt.hist(df['Genre'], bins=20)
plt.xticks(rotation=90)
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

### 6. Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)

```python
count = df['Released_Year'].value_counts()
plt.xticks(rotation=90)
plt.scatter(count.index, count.values)
```
